```python
# findall()  &  split() parse strings and return chunks 
string = "you are amazing . you can do it . Nobody can beat you"

re.split("you",string)
```

## Patterns and Character Classes

```python
# Grades across a semester 
grades = "AABBBCABCBCACB"
# Find all B's
re.findall("B",grades)
```

```python
# Find all A's OR B's
re.findall("[AB]",grades)
# This is called SET OPERATOR 
```

```python
# Can also include ranges like [a-z]
# All AB's OR AC's
re.findall("[A][B-C]",grades)
```

```python
# This can also be written using pipe operator 
re.findall("AB|AC",grades)
```

```python
# '^' can be used to negate results
# for all grades that are not A
# This can also be written using pipe operator 
re.findall("[^A]",grades)
```

```python
re.findall("^[^A]",grades)
```
