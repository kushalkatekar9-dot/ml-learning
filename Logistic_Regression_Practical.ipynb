```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df["species"].unique()
```

```python
df.isna().sum()
```

```python
df = df[df['species'] != 'setosa']
df.head()virginica
```

```python
# MAP
df['species'] = df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.head()
```

```python
# SPLIT DATASET INTO INDEPENDANT AND DEPENDDANT FEATURES 
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,7,21,53,64,25,63,79],'max_iter':[100,111,234,666,300,546]}
```

```python
classifier_regressor = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=20)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
## prediction
y_pred = classifier_regressor.predict(X_test)
```

```python
## accuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
