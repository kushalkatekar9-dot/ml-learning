## Patterns and Character Classes

```python
# Grades across a semester 
grades = "AABBBCABCBCACB"
# Find all B's
re.findall("B",grades)
```

```python
# Find all A's OR B's
re.findall("[AB]",grades)
# This is called SET OPERATOR 
```

```python
# Can also include ranges like [a-z]
# All AB's OR AC's
re.findall("[A][B-C]",grades)
```

```python
# This can also be written using pipe operator 
re.findall("AB|AC",grades)
```

```python
# '^' can be used to negate results
# for all grades that are not A
# This can also be written using pipe operator 
re.findall("[^A]",grades)
```

```python
re.findall("^[^A]",grades)
```

```python
# This gives an empty list   ** because the regex says that we want to match any value at the beginning of the string
# which is not an A. Our string though starts with an A, so there is no match found. And remember when you are
# using the set operator you are doing character-based matching. So you are matching individual characters in
# an OR method.
```

## Quantifiers 

```python
# Quantifiers are the number of times you want a pattern to be matched in order to match
# Most basic --> e{m,n} e is the expression/character we wan to match
# m is the min number of times you want it matched and n is the max number of times 
grades = "AABBBCABCBCACB"
re.findall("A{1,10}",grades)
```

```python
# if you just have one number in the braces, it's considered to be both m and n
#if we don't include a quantifier then the default is {1,1}
# It's important to note that the regex quantifier syntax does not allow you to deviate from the {m,n}
# pattern. In particular, if you have an extra space in between the braces you'll get an empty result
re.findall("A{2, 2}",grades)
```

```python
# Using this, we could find a decreasing trend in a student's grades
grades = "AABBBCABCBCACB"
re.findall("A{1,10}B{1,10}C{1,10}",grades)
```
